```python
df.drop(columns=['Data'],inplace=True)
```

```python
import re

df['Time_Hour'] = df['Time'].apply(lambda x: re.search(r'^\d+', x).group(0)).astype(np.int)
df['Time_Minute'] = df['Time'].apply(lambda x: re.search(r'(?<=:)\d+(?=:)', x).group(0)).astype(np.int)
df['Time_Second'] = df['Time'].apply(lambda x: re.search(r'\d+$', x).group(0)).astype(np.int)

df = df.drop('Time', axis=1)
```

```python
df.head()
```

```python
df['SunriseHour'] = df['TimeSunRise'].apply(lambda x: re.search(r'^\d+', x).group(0)).astype(np.int)
df['SunriseMinute'] = df['TimeSunRise'].apply(lambda x: re.search(r'(?<=:)\d+(?=:)', x).group(0)).astype(np.int)

df['SunsetHour'] = df['TimeSunSet'].apply(lambda x: re.search(r'^\d+', x).group(0)).astype(np.int)
df['SunsetMinute'] = df['TimeSunSet'].apply(lambda x: re.search(r'(?<=:)\d+(?=:)', x).group(0)).astype(np.int)

df = df.drop(['TimeSunRise', 'TimeSunSet'], axis=1)
```

```python
df.drop(columns=['SunriseHour'],inplace=True)
```

```python
x=df.drop(['Radiation'],axis='columns')
y=df.Radiation.values
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=300)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rnd=RandomForestClassifier()
rnd = RandomForestRegressor(max_depth=25, random_state=3)
rnd.fit(xtrain, ytrain)
```

```python
rnd.score(xtrain,ytrain)
```

```python
rnd.score(xtest,ytest)
```
